```
import pandas as pd
from pandas.io.json import json_normalize 
import matplotlib.pylab as plt
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
# Below import will allow multiple outputs per cell
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = 'all'
```

```
df = pd.read_json('AML_Azerbaijan_data.json')
df.head(5)
```

```
# Convert that column into a 'normalized' DataFrame with json_normalize
df = json_normalize(df['data'])
```

```
df.head()
```

```
pd.options.display.max_columns = 20
```

```
pd.options.display.float_format = "{:.2f}".format
df.describe()
```

```
df.describe(include=['O', 'bool'])
```

```
#amount of currency in the transactions
df.groupby(['amount_orig_currency'])['amount_orig_currency'].count()
```

```
df[['payer_account', 'beneficiary_account']].agg(['nunique', 'count']).T
```

```
df['amount_eur'] = df['amount_eur'].str.replace('$','')
df['amount_eur'] = df['amount_eur'].str.replace(',','')
df['amount_eur'] = df['amount_eur'].astype(float)
```

```
df['amount_orig'] = df['amount_orig'].abs()
```

```
df['beneficiary_jurisdiction'].unique()
```

```
df['beneficiary_jurisdiction'] = df['beneficiary_jurisdiction'].replace('CN\n', 'CN')
df['beneficiary_jurisdiction'] = df['beneficiary_jurisdiction'].replace('UNKNOWN', np.nan)
df['beneficiary_jurisdiction'].unique()
```

```
df['payer_bank_country'] = df['payer_bank_country'].fillna(np.nan)
df['payer_bank_country'].unique()
```

```
df.groupby('date')['amount_usd'].sum().nlargest(5)
```

```
df['date'] = pd.to_datetime(df['date'])
dk_ = pd.Series(df['amount_usd'].values, index=df.date)
dk_.plot(legend=False, figsize=[20, 10])
plt.xlabel('Timeline')  
plt.ylabel('Transaction amounts (in 0000000s $)') 
plt.tight_layout()
plt.show()
```

```
df['transaction_direction'] = np.where(
    np.logical_and(df['beneficiary_core']==True,df['payer_core']==True),'Internal Routing',np.where(
        df['payer_core']==True,'Outgoing Transaction','Incoming Transaction'))
df['transaction_direction'].value_counts()
```

```
#counted transactions.
dup_trans = df[df.duplicated(subset=['amount_usd','date','payer_name_norm','beneficiary_name_norm'])]
#There seem to be 342 such transactions which have been recorded twice
df['duplicated_transaction'] = np.where(df['id'].isin(dup_trans['id'].values),1,0)
df[df['duplicated_transaction']==1]
```

```
#focus on where the money goes (beneficiary) when there is no payer account

series_to_plot = df.loc[df['payer_account'].isna(), 'beneficiary_name_norm'].value_counts()

series_to_plot
```

```
sns.barplot(series_to_plot.index, series_to_plot.values)
```

```
top_10_payers = df.groupby(['payer_name_norm'])['amount_eur'].sum().reset_index().sort_values(by='amount_eur', ascending=False).head(10)
top_10_payers
```

```
sns.set(rc={'figure.figsize': (25,9)})
sns.barplot(data=top_10_payers, x='payer_name_norm', y='amount_eur')
```
